```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
```

```python

df = pd.read_csv("Saumons.csv")

df_clean = df[['Method', 'Month', 'Region']].dropna()
```

```python
contingence_mr = pd.crosstab(df_clean['Method'], df_clean['Region'])

chi2, p, dof, expected = chi2_contingency(contingence_mr)

plt.figure(figsize=(10, 6))
sns.heatmap(contingence_mr, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Table de contingence : Méthode × Région")
plt.xlabel("Région")
plt.ylabel("Méthode")
plt.tight_layout()
plt.show()

```

```python
print("Statistique du test khi² :", chi2)
print("p-valeur :", p)

if p < 0.05:
    print("➤ Il existe une dépendance significative entre la méthode de pêche et la région.")
else:
    print("➤ Aucune dépendance significative n'a été détectée entre les deux variables.")
```

On observe des différences marquées dans l’utilisation des méthodes selon les régions :

Fixed Engine: Retained domine dans toutes les régions, surtout dans Solway, North East et Moray Firth.

Net and Coble: Retained est aussi bien représentée, notamment dans East et Moray Firth.

Fixed Engine: Released et Net and Coble: Released sont très peu utilisées (presque absentes).

Test du χ² d’indépendance
Statistique χ² : 1368.56
p-valeur : 1.09e-271 (≈ 0)


Cela signifie qu’il y a une dépendance statistique très significative entre la méthode de pêche et la région :
le choix de la méthode varie selon la région.


```python
contingence_mm = pd.crosstab(df_clean['Method'], df_clean['Month'])

chi2_mm, p_mm, dof_mm, expected_mm = chi2_contingency(contingence_mm)

plt.figure(figsize=(10, 6))
sns.heatmap(contingence_mm, annot=True, fmt="d", cmap="YlOrRd")
plt.title("Table de contingence : Méthode × Mois")
plt.xlabel("Mois")
plt.ylabel("Méthode")
plt.tight_layout()
plt.show(), chi2_mm, p_mm
```
