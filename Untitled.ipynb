```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
```

```python

df = pd.read_csv("Saumons.csv")

df_clean = df[['Method', 'Month', 'Region']].dropna()
```

```python
contingence_mr = pd.crosstab(df_clean['Method'], df_clean['Region'])

chi2, p, dof, expected = chi2_contingency(contingence_mr)

plt.figure(figsize=(10, 6))
sns.heatmap(contingence_mr, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Table de contingence : Mthode  Rgion")
plt.xlabel("Rgion")
plt.ylabel("Mthode")
plt.tight_layout()
plt.show()

```

```python
print("Statistique du test khi :", chi2)
print("p-valeur :", p)

if p < 0.05:
    print(" Il existe une dpendance significative entre la mthode de pche et la rgion.")
else:
    print(" Aucune dpendance significative n'a t dtecte entre les deux variables.")
```

On observe des diffrences marques dans lutilisation des mthodes selon les rgions :

Fixed Engine: Retained domine dans toutes les rgions, surtout dans Solway, North East et Moray Firth.

Net and Coble: Retained est aussi bien reprsente, notamment dans East et Moray Firth.

Fixed Engine: Released et Net and Coble: Released sont trs peu utilises (presque absentes).
