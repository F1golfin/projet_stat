```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des données
saumons = pd.read_csv("saumons.csv", sep=',')

# Vérifier les premières lignes du DataFrame
print(saumons.head())
print(saumons.columns)
```

```python
# Importation des données
Method= saumons["Method"];
# Nombre d'occurrences par catégorie 
saumons["Method"].value_counts()  
```

```python
# On trace ensuite un histogramme pour visualiser la répartition 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,16))
ax = sns.countplot(x=saumons["Method"], order=saumons["Method"].value_counts().index)
plt.xlabel("Méthodes de captures")
plt.ylabel("Nombre d'observations")
plt.title("Répartition des méthodes de captures")
plt.show()
```

```python
#Création d'un histogramme avec les fréquences cumulées
counts = saumons["Method"].value_counts()
cumulative_frequencies = counts.cumsum() / counts.sum()  # Normalisation en pourcentage
# Création du graphique
plt.figure(figsize=(12, 6))
sns.barplot(x=cumulative_frequencies.index, y=cumulative_frequencies.values)
plt.xlabel("Méthodes de capture")
plt.ylabel("Fréquence cumulée")
plt.title("Fréquence cumulée des méthodes de capture")
plt.show()
```

```python
# Importation des données
Region= saumons["Region"];
```

```python
# Nombre d'occurrences par catégorie 
saumons["Region"].value_counts() 
```

```python
# Affichage d'un diagramme circulaire
saumons["Region"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(15, 20))
plt.title("Répartition des différentes régions")
plt.ylabel("")
plt.show()
```

Comment sont réparties les différentes méthodes de pêche en fonction des régions ? Peut-on observer des méthodes plus présentes dans certains régions ?

```python
df = pd.read_csv("saumons.csv")
table = pd.crosstab(df["Method"], df["Region"])
```

```python
plt.figure(figsize=(10, 6))  # Taille du graphique
sns.heatmap(table, annot=True, cmap="BuPu", fmt=".0f",linewidths=0.5)  # annot=True pour afficher les nombres

plt.title("Fréquence d'observation selon la méthode et la région")
plt.ylabel("Méthode de pêche")
plt.xlabel("Région")
plt.tight_layout()
plt.show()
```

Dans cette heatmap nous pouvons observer plusieurs informations intéressantes :

- Les méthodes "Fixed Engine: Release" et "Net and Coble: Release" sont très peu utilisées en général avec un effectif cumulé total inférieur à 100.
- Les méthodes "Fixed Engine: Retained" et "Net and Coble: Retained" sont les plus utilisées avec une majorité écrasante des effectifs concentrée dans ces méthodes.

Nous pouvons en conclure que les méthodes de type "Retained" sont celles favorisées pour pêcher le saumon.

Ensuite, nous pouvons observer une répartition de ces deux méthodes assez homogène en fonction des régions avec des effectifs compris entre 2000 et 3000 en moyenne par région.
Nous pouvons tout de même constater que la méthode "Fixed Engine: Retained" semble plus utilisée que la "Net and Coble: Retained" avec des "cases" de la heatmap plus sombres.

Par ailleurs, La méthode "Fixed Engine: Retained" semble particulièrement appréciée et utilisée dans la région "North East" et la région "Solway".

Finalement, il semble que les régions "Orkney" et "Outer Hebrides" ne sont pas des régions très prisées dans la pêche de saumon avec des effectifs particulièrement bas, toutes méthodes confondues.

Ainsi, cette analyse met en lumière une utilisation particulièrement conséquente des méthodes de type "Retained" pour pêcher le saumon, toutes régions confondues.
Et parmis ces méthodes, la méthode de type "Fixed Engine: Retained" est la plus prisée dans la majorité des régions à l'exception de "Outer Hebrides" et de "East" qui ont une utilisation de la méthode "Net and Coble: Retained" légèrement plus élevée.
