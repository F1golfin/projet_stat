```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des données
saumons = pd.read_csv("saumons.csv", sep=',')

# Vérifier les premières lignes du DataFrame
print(saumons.head())
print(saumons.columns)
```

Comment est réparti le poids de Finnock pêché en fonction de sa région ? Peut-on observer des regions particulières selon ce critère ?

```python
df = pd.read_csv("saumons.csv")
table = pd.crosstab(df["Finnock weight (kg)"], df["Region"])
```

```python
plt.figure(figsize=(10, 6))  # Taille du graphique
sns.heatmap(table, annot=True, cmap="BuPu", fmt=".0f",linewidths=0.5)  # annot=True pour afficher les nombres

plt.title("Fréquence d'observation selon le poids et la région")
plt.ylabel("Poids de Finnock(en kg)")
plt.xlabel("Région")
plt.tight_layout()
plt.show()
```

Cette heatmap est très parlante sur deux aspect : 
- Le poids de finnock pêché est particulièrement concentré entre 0g et 200g avec quasiment la totalité des effectifs dans cette fourchette de valeurs
- Les régions Orkney et Outer Hebrides sont très peu utilisées pour pêcher le saumon, en témoignent les cases très "claires" dans la heatmap.

On remarque tout de même que les effectifs se situent entre 2000 et 4000 pour chacune des autres régions avec un maximum de 3981 atteint dans la région North East.

On observe également des valeurs extrêmes atteintes dans certaines régions comme un Finnock d'environ 5,7kg pêché dans la région "West Coast" ce qui est imrpessionnant lorsqu'on voit que la moyenne tourne entre 0 et 200g.

Finalement, cette analyse bivariée ne se montre pas très parlante car les poids sont particulièrement concentrés dans le même point et la répartition des effectifs selon les régions est à peu près homogène avec quelques extrêmes.
